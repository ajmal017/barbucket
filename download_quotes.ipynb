```python
import fix_yahoo_finance as yf
from tqdm import tqdm_notebook as tqdm
import time
import json
import os
from time import strftime, localtime
```

```python
symbols = []

file = open('xetra_etf_symbols.txt','r') 
for line in file:
    symbols.append(line.rstrip())
file.close()

# symbols
```

```python
interval = '1d'
start = '2000-01-01'
```

```python
for symbol in tqdm(symbols):
    ticker = yf.Ticker(symbol)

    file_path = f'data\{symbol}.csv'
    file_exists = os.path.isfile(file_path)
    file_complete = False

    if file_exists:
        file_mdate = os.path.getmtime(file_path)
        if (strftime('%x', localtime(file_mdate)) == strftime('%x', localtime())):
            file_complete = True

    if not file_complete:
        data = ticker.history(interval=interval, start=start, period=None)
        data.to_csv(file_path, index='Date')
        time.sleep(1)
```

```python
for symbol in tqdm(symbols):
    ticker = yf.Ticker(symbol)
    
    file_path = f'data\{symbol}.json'
    file_exists = os.path.isfile(file_path)
    file_complete = False    
    
    if file_exists:
        file_mdate = os.path.getmtime(file_path)
        if (strftime('%x', localtime(file_mdate)) == strftime('%x', localtime())):
            file_complete = True

    if not file_complete:    
        data = ticker.info
        file = open(file_path, 'w')
        json.dump(data, file)
        file.close()
        time.sleep(1)
```
