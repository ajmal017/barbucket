```python
symbols = []

file = open('xetra_etf_symbols.txt','r') 
for line in file:
    symbols.append(line.rstrip())
file.close()

symbols
```

```python
import fix_yahoo_finance as yf
from tqdm import tqdm_notebook as tqdm
import time
import json
```

```python
msft = yf.Ticker("MSFT")

# get stock info
msft.info
```

```python
hist = msft.history(period="max")
hist
```

```python
interval = '1d'
start = '2000-01-01'
```

```python
for symbol in tqdm(symbols):
    ticker = yf.Ticker(symbol)
    data = ticker.history(interval=interval, start=start, period=None)
    data.to_csv(f'data\{symbol}.csv', index='Date')
    time.sleep(1)
```

```python
for symbol in tqdm(symbols):
    ticker = yf.Ticker(symbol)
    data = ticker.info
    file = open(f'data\{symbol}.json', 'w')
    json.dump(data, file)
    time.sleep(1)
```
